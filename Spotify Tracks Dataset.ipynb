```python
import pandas as pd

# definicion del path
path = "/content/drive/MyDrive/Colab Notebooks/Proyecto Final Diseno Algoritmos/dataset.csv"
file = ""

df_spotify = pd.read_csv(path, encoding= 'utf-8')

df_spotify.head(5)
```

#2) Conocer los tipos de datos de las columnas para verificar que todo este correcto y además, eliminar canciones duplicadas, ya que este dataset cuenta con un nombre para la canción y la pista, por lo que duplica los nombres.

```python
print(df_spotify.shape)
print(df_spotify.size)
print(df_spotify.dtypes)
```

###Una vez corroborada la información, Procedemos a limpiar el dataset original y creando uno nuevo.

```python
df_spotifyClean = df_spotify.drop_duplicates(subset='track_name')

print(df_spotifyClean.shape)
print(df_spotifyClean.size)

df_spotifyClean.head(10)
```

#3) Comenzaremos analizando las canciones mas populares y ordenandolas de forma ascendente.

###La columna popularidad, solo remarca que tan escuchada es una cancion, mientras mas reproducciones tenga mas popularidad tendrá.

####Analizaremos las canciones con popularidad de 90 - 100, ordenandolas de forma descendente. y guardandolas en un nuevo dataframe. (df_popular)

```python
df_popular = df_spotifyClean[ (df_spotifyClean['popularity'] >= 90 ) & (df_spotifyClean['popularity'] <=100)].sort_values('popularity', ascending=False)

print(df_popular.shape)

df_popular.head(40)
```

#4) Cual es la tonalidad musical mas tocada en spotify (Las mas populares).

##Para este caso haremos uso de la columna key, que nos dice la clave musical sobre la que esta la cancion. Además, graficaremos para tener una visualización mas clara del análisis.

```python
df_popular.groupby(['key']).count().plot(y='track_name', kind='bar', figsize=(30,15))
```

##En este caso, las tonalidades van marcadas por numeros, donde 0=DO, 1=DO#, 2=RE, 3=RE#, Etc.

##Como conclusión, las tonalidades mas populares son RE, G y LA#

#5) ¿Cual es el género de música mas popular? (Entre lo mas escuchado)

```python
df_popular.groupby(['track_genre']).count().plot(y='artists', kind='bar', figsize=(30,15))
```

#6) ¿Cuantas canciones explicitas hay? son mas populares?

```python
df_popular.groupby('explicit').count().plot(y='track_name', kind='bar', figsize=(20,10))
```

#7) Implementacion de funciones.

###Creamos una función que nos ayude a encontrar canciones populares de un artista en específico

###Creamos una función capáz de devolvernos canciones populares de un género en un rango de popularidad en específico

```python
def popularSong(df, art, rInf, rSup):
  return df[ (df['popularity'] >= rInf) & (df['popularity'] <= rSup) & (df['artists'] == art)]
```

```python
popularSong(df_spotifyClean, 'AC/DC', 80, 100)
```

```python
def popularGenre(df, genre, rInf, rSup):
  return df[ (df['popularity'] >= rInf) & (df['popularity'] <= rSup) & (df['track_genre'] == genre)]
```

```python
popularGenre(df_spotifyClean, 'latin', 70, 100)
```
